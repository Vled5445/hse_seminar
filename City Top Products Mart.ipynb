```scala
%spark.pyspark
from pyspark.sql import functions as F
from pyspark.sql.window import Window

users = spark.createDataFrame(
    [
        ("u1", "Berlin"),
        ("u2", "Berlin"),
        ("u3", "Munich"),
        ("u4", "Hamburg"),
    ],
    ["user_id", "city"]
)

orders = spark.createDataFrame(
    [
        ("o1", "u1", "p1", 2, 10.0),
        ("o2", "u1", "p2", 1, 30.0),
        ("o3", "u2", "p1", 1, 10.0),
        ("o4", "u2", "p3", 5, 7.0),
        ("o5", "u3", "p2", 3, 30.0),
        ("o6", "u3", "p3", 1, 7.0),
        ("o7", "u4", "p1", 10, 10.0),
    ],
    ["order_id", "user_id", "product_id", "qty", "price"]
)

products = spark.createDataFrame(
    [
        ("p1", "Ring VOLA"),
        ("p2", "Ring POROG"),
        ("p3", "Ring TISHINA"),
    ],
    ["product_id", "product_name"]
)

print("Данные созданы!")
users.show()
orders.show()
products.show()
```

```scala
%spark.pyspark
orders_with_revenue = orders.withColumn("revenue", F.col("qty") * F.col("price"))

enriched_orders = (
    orders_with_revenue
    .join(users, on="user_id", how="inner")
    .join(products, on="product_id", how="inner")
    .select("order_id", "city", "product_id", "product_name", "qty", "price", "revenue")
)

print("Объединенные данные с revenue:")
enriched_orders.show(truncate=False)
```

```scala
%spark.pyspark

city_product_metrics = (
    enriched_orders
    .groupBy("city", "product_id", "product_name")
    .agg(
        F.count("order_id").alias("orders_cnt"),
        F.sum("qty").alias("qty_sum"),
        F.sum("revenue").alias("revenue_sum")
    )
)

print("Метрики по городам и продуктам:")
city_product_metrics.orderBy("city", F.col("revenue_sum").desc()).show(truncate=False)
```

```scala
%spark.pyspark
window_spec = Window.partitionBy("city").orderBy(F.col("revenue_sum").desc())


mart_city_top_products = (
    city_product_metrics
    .withColumn("rank", F.row_number().over(window_spec))
    .filter(F.col("rank") <= 2)
    .select("city", "product_id", "product_name", "orders_cnt", "qty_sum", "revenue_sum", "rank")
    .orderBy("city", "rank")
)

print("Топ-2 продукта по сумме выручки в каждом городе:")
mart_city_top_products.show(truncate=False)
```

```scala
%spark.pyspark

s3_path = "s3a://my-pyspark-bucket-123/mart_city_top_products/"
mart_city_top_products.write.mode("overwrite").parquet(s3_path)
print(f"Данные сохранены в S3: {s3_path}")


local_path = "/tmp/sandbox_zeppelin/mart_city_top_products/"
mart_city_top_products.write.mode("overwrite").parquet(local_path)
print(f"Данные сохранены локально: {local_path}")
```

```scala
%spark.pyspark

print("Читаем данные из S3")
s3_mart = spark.read.parquet("s3a://my-pyspark-bucket-123/mart_city_top_products/")
print("Схема прочитанных данных:")
s3_mart.printSchema()
print("Топ-результат из S3:")
s3_mart.orderBy("city", "rank").show(truncate=False)

print("Читаем данные из локального пути")
local_mart = spark.read.parquet("/tmp/sandbox_zeppelin/mart_city_top_products/")
print("Топ-результат из локального хранилища:")
local_mart.orderBy("city", "rank").show(truncate=False)
```
